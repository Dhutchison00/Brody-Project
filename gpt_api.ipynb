```python
def openImage(img):

    # Get the image from img
    img = cv2.imread(img)

    # Convert the image from RGB to HSV
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    
    # Shrink both images
    img = cv2.resize(img, (0,0), fx=0.5, fy=0.5)
    hsv = cv2.resize(hsv, (0,0), fx=0.5, fy=0.5)

    # Only include pixels in the range of the color
    mask = cv2.inRange(hsv, (0, 8, 140), (160, 250, 220))

    
    # Apply the mask on hsv
    hsv_mask = cv2.bitwise_and(img, img, mask=mask)

    cv2.imshow('hsv',img)
    cv2.imshow('mask',hsv_mask)

    # On mouse click, print the HSV value of the pixel
    def mouse_callback(event, x, y, flags, params):
        if event == cv2.EVENT_LBUTTONDOWN:
            print(hsv[y,x])
    
    # Set the mouse callback function
    cv2.setMouseCallback('hsv', mouse_callback)

    cv2.waitKey(0)
    cv2.destroyAllWindows()

openImage('car.jpeg')

#https://www.tutorialspoint.com/how-to-mask-an-image-in-opencv-python
# have brody put a green cube in the middle of the img.
```

```python
def faceDetection(img):

    # Get the image from img
    img = cv2.imread(img)

    # Shrink image
    img = cv2.resize(img, (0,0), fx=0.5, fy=0.5)
    
    faces = DeepFace.analyze(img, actions=['age'])

    for face in faces:
        location = face['region']  
        x = location['x']
        y = location['y']
        w = location['w']
        h = location['h']

        cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0), 2)
        cv2.putText(img, str(face['age']), (x,y), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)

    

    cv2.imshow('face',img)

    cv2.waitKey(0)
    cv2.destroyAllWindows()

faceDetection('face.jpg')

#talk about his code and how enhanced loops are better
```

```python
def openWebcamWindow():

    # Set up webcam capture
    vid = cv2.VideoCapture(0)


    while True:
        # Get current frame from webcam
        _, frame = vid.read()

        faces = DeepFace.analyze(frame, actions=['gender'], enforce_detection=False, detector_backend='opencv')

        for face in faces:
            location = face['region']  
            x = location['x']
            y = location['y']
            w = location['w']
            h = location['h']

            cv2.rectangle(frame, (x,y), (x+w,y+h), (0,255,0), 2)
            cv2.putText(frame, str(face['dominant_gender']), (x,y), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)

        # Display current frame
        cv2.imshow("Webcam", frame)

        # Press q to quit
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    vid.release()
    cv2.destroyAllWindows()

openWebcamWindow()

# have brody look up and try to figure out by himself how to put a green square in the middle of the screen
```
